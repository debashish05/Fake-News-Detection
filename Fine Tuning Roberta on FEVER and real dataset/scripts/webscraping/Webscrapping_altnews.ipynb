```python
url="https://www.altnews.in/"
```

```python
import os
from selenium.webdriver.common.keys import Keys
import time
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
```

```python
driver = webdriver.Chrome(executable_path='chromedriver-3')
```

```python
driver.get(url)
```

```python
screen_height = driver.execute_script("return window.screen.height;")   
start = time.time()
i=1
t1=0
while True:
    driver.execute_script("window.scrollTo(0, {screen_height}*{i});".format(screen_height=screen_height, i=i))  
    i += 1
    time.sleep(2)
    end = time.time()
    t2 = round(end - start)
    if round(t2/60)!=t1:
        t1=round(t2/60)
        print(t1)
    if t2 > 7200:
        break
```

```python
import os
if os.path.exists("DS.html"):
  os.remove("DS.html")
```

```python
file = open('DS.html', 'w',encoding='utf-8')
file.write(driver.page_source)
file.close()

driver.close()
```

```python
html = open('DS.html','r',encoding='utf-8')
```

```python
soup = BeautifulSoup(html,'html.parser')
```

```python
print(soup)
```

```python
print(soup.prettify())
```

```python
soup.title
```

```python
title = [i.text.strip() for i in soup.select('h4.entry-title.pbs_e-t.xs__h6.sm__h5')]
title
```

```python
link=[]
for links in soup.select('h4.entry-title.pbs_e-t.xs__h6.sm__h5'):
        link.append(links.find('a').get('href'))
```

```python
link
```

```python
len(link)
```

```python
content=[]
for i in range (len(link)):
    p=requests.get(link[i]).text
#     time.sleep(1)
    s=BeautifulSoup(p,"html.parser")
#     time.sleep(1)
    content.append(s.select('p')[0].text)
#     time.sleep(2)
```

```python
content
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({"article link":link,"article title":title,"article content":content})
```

```python
df.to_csv("altnews2.csv")
```
