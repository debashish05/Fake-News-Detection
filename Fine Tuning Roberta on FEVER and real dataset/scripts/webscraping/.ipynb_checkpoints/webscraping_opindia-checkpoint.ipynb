```python
# Importing BeautifulSoup
from bs4 import BeautifulSoup
```

```python
# Importing pandas for creating dataframe
import pandas as pd
```

```python
# Importing requests
import requests 
```

```python
# Extracting links for all pages of opindia page
page_url ='https://www.opindia.com/category/fact-check/page/'
urls=[]
for i in range(1,138):
  word=page_url+str(i)
  urls.append(word)


# Extracting soup for all the above page url's
soup_arr=[];
for url in urls:
  html_extract = requests.get(url).text
  soup=  BeautifulSoup(html_extract,'html.parser')
  soup_arr.append(soup)
len(soup_arr)
```

```python
cnt=0
```

```python
# To store the links of every individual article
all_url = []
```

```python
# Extracting links for all articles from the soups we created before and storing the links in all_url
cnt=1
for soup in soup_arr:
    for itr in soup.find_all('div',id="tdi_119" , class_='td_block_inner tdb-block-inner td-fix-index'):
        for ul in itr.find_all('div', class_='td-module-container td-category-pos-image'):
            for li in ul.find_all('h3', class_='entry-title td-module-title'):
                print(cnt)
                a = li.find('a',href=True)
                print(a['href'])
                print('\n')
                cnt=cnt+1
                all_url.append(a['href'])
            
        
        
        
       
   
        
        
        
        
        
        
```

```python
# To store all the data in dataframe
df=[]
```

```python
cnt=1
```

```python
# Extracting the required content ie. article link, article heading and article content from each URL stored in all_url
new_row = []
for link in all_url:
    
    url =link
    html_extract = requests.get(url).text
    #time.sleep(2)

    soup =  BeautifulSoup(html_extract,'html.parser')
    new_row.append(url)
    

    for title in soup.find_all('h1',class_='tdb-title-text'):
        #print(title.get_text())
        #print('\n')
        continue
    new_row.append(title.get_text())
        
        
    text=""
    for itr in soup.find_all('div' , class_='tdb-block-inner td-fix-index'):   
        for ul in itr.find_all('p'):
        
            text=text+(ul.get_text())
            #print(ul.get_text()) 
    new_row.append(text)
    #print('\n')
    
    # Appending each new row in dataframe (2D-array)
    df.append(new_row)
    
    new_row=[]
    
    
    
    
    
```

```python
# Creating final dataframe
df_main = pd.DataFrame(df,columns=['Article_link','Article_title','Article_content'])
```
