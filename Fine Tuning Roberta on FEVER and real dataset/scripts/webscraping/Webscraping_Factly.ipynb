## Automating Script for Dataframe

```python
stud='https://factly.in/category/english/page/{}/'
```

```python
stud.format(5)
```

```python
def headlines(soup):
    news_headline = [i.text.strip() for i in soup.select('h2.post-title')]
    news_headline = news_headline[1:]
    return pd.DataFrame({"Headline":news_headline})
```

```python
def content(soup):
    urls =[]
    contents =[]
    for i in soup.select('h2.post-title'):
        urls.append(i.a['href'])
    urls = urls[1:]
    for i in tqdm(range(len(urls))): 
        browser.get(urls[i])
        time.sleep(2)
        html = browser.page_source
        bsobj=BeautifulSoup(html,'html.parser')
        contents.append([i.text.strip() for i in bsobj.select('p.selectionShareable')])
    return pd.DataFrame({"Content":contents,"Link":urls})
```

```python
# 125 pages from factly.in
tables=[]
tables_1 =[]

for i in tqdm(range(1,126)):
    url_pages = stud.format(i)
    browser.get(url_pages)
    time.sleep(2)
    html = browser.page_source
    soup = BeautifulSoup(html,"html.parser")
    tables.append(headlines(soup))
    tables_1.append(content(soup))
```

```python
df = pd.concat(tables,axis=0)
df_1 = pd.concat(tables_1,axis=0)
final_df = pd.concat([df,df_1],axis=1)
```

```python
data = final_df.reset_index(drop=True)
```
