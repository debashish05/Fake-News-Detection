###  Subtask 2: Getting headline and body text from all links and pushing it into a dataframe

```python
stud = 'https://www.boomlive.in{}'
```

```python
df = pd.DataFrame(columns=['Article Link', 'Article Title', 'Article Content'])
```

```python
for i in tqdm(range(len(links))):
    url = stud.format(links[i])
    html_good = requests.get(url).text
    time.sleep(1)
    soup = BeautifulSoup(html_good, "html.parser")
    
    #Headline
    headline_array = soup.find_all('h1',class_='entry-title mb-10')
    if headline_array:
        headline = str(soup.find_all('h1',class_='entry-title mb-10')[0].string)
    
    #Content
    line = soup.find_all('p')
    if(line):
        content = ''
        for i in range(len(line)-1):
            res = re.sub(r'<.*?>', '', str(line[i]))
            if isEnglish(res):
                content = content + ' ' + res
    
    #Adding data to dataframe
    if headline_array and line:
        data = {'Article Link': url, 'Article Title': headline, 'Article Content': content}
        df = df.append(data, ignore_index = True)
    
```
